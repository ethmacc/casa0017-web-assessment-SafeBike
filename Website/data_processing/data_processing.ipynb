```python
import geopandas as gpd
import pandas as pd
import os
import matplotlib as mpl
from matplotlib import cm
```

```python
MPS_crime = pd.read_csv('MPS LSOA Level Crime (most recent 24 months).csv')
MPS_crime
```

```python
bike_theft = MPS_crime[MPS_crime['Minor Category'] == 'BICYCLE THEFT'].reset_index(drop=True)
#bike_theft['total'] =  bike_theft.iloc[:, 6:].sum(axis=1)
#bike_theft = bike_theft.rename(columns={'LSOA Code':'lsoa21cd', '202409':'most_recent'})
#bike_theft_short = bike_theft[['lsoa21cd', 'most_recent', 'total']]
bike_theft_short = bike_theft.drop(columns=['Major Category', 'Minor Category'])
#bike_theft_short.to_csv('bike_theft.csv', index=False)
```

```python
total = bike_theft_short.iloc[:, 3:].sum(axis=1)
#monthly = bike_theft_short.iloc[:, 3:].values.tolist()
bike_theft_compact = bike_theft_short#.iloc[:, :3]
#bike_theft_compact['Monthly'] = monthly
bike_theft_compact['Total'] = total
bike_theft_compact
```

```python
shape_files = [f for f in os.listdir('LB_shp/') if '.shp' in f]

dat_ls = []
for file in shape_files:
    temp = gpd.read_file(f'LB_shp/{file}')
    dat_ls.append(temp)

lsoa2021 = pd.concat(dat_ls)
lsoa2021.geometry = lsoa2021.geometry.to_crs(epsg=4326)
lsoa2021 = lsoa2021.reset_index(drop=True)
```

```python
lsoa2021.geometry = lsoa2021.geometry.simplify(0.00001)
```

```python
lsoa2021
```

```python
merged = lsoa2021.merge(bike_theft_compact, left_on='lsoa21cd', right_on='LSOA Code', how='left')
merged = merged.drop(columns=['LSOA Code', 'LSOA Name', 'msoa21cd', 'msoa21nm', 'Borough'])
merged = merged.rename(columns={'lsoa21cd':'LSOA Code', 'lsoa21nm':'LSOA Name', 'lad22cd':'Borough', 'lad22nm':'Borough Name'})
#merged = merged.fillna('No data')

#def assign_color_rgb(times):
 #   output_col = scalarMap.to_rgba(times)
  #  return f'rgb({int(output_col[0] * 255)}, {int(output_col[1] * 255)}, {int(output_col[2] * 255)})' 

#merged['color'] = merged['Total'].apply(assign_color_rgb)

merged
```

## Plotting

```python
import altair as alt
import datetime as dt
```

```python
to_plot = merged.drop(columns=['geometry', 'Total'])
```

```python
totals = to_plot.iloc[:,4:].sum().T
```

```python
totals = pd.DataFrame(totals)
```

```python
totals['Months'] = pd.to_datetime(totals.index.str[4:] + '-' + totals.index.str[0:4], format='%m-%Y')
```

```python
totals = totals.rename(columns={0:'Total thefts'})
```

```python
# Adapted from https://altair-viz.github.io/gallery/multiline_tooltip.html

def plot_theft_chart(totals):
    nearest = alt.selection_point(nearest=True, on="pointerover", fields=["Months"], empty=False)

    line = alt.Chart(totals).mark_line().encode(
        x=alt.Y('Months:T',
                scale=alt.Scale(domain=[dt.datetime(2022, 9, 1),dt.datetime(2024, 10, 1)]),
                axis=alt.Axis(format="%B %Y", tickCount=24, labelAngle=80),
               ),
        y=alt.X('Total thefts')
    )

    selectors = alt.Chart(totals).mark_point().encode(x='Months', opacity=alt.value(0)).add_params(nearest)
    
    when_near = alt.when(nearest)
    
    points = line.mark_point().encode(opacity=when_near.then(alt.value(1)).otherwise(alt.value(0)) )
    
    text = line.mark_text(align="left", dx=5, dy=-5).encode(text=when_near.then("Total thefts").otherwise(alt.value(" ")) )
    
    rules = alt.Chart(totals).mark_rule(color="gray").encode(x="Months").transform_filter(nearest)
    
    final_chart = alt.layer(line, selectors, points, rules, text).properties(width=600, height=300)

    return final_chart
```

```python
total_chart = plot_theft_chart(totals)
total_chart.show()
```

```python
total_chart.save('total_thefts.html')
```

```python
for borough in to_plot['Borough Name'].unique():
    borough_plot = to_plot[to_plot['Borough Name'] == borough]
    borough_total = borough_plot.iloc[:,4:].sum().T
    borough_total = pd.DataFrame(borough_total)
    borough_total['Months'] = pd.to_datetime(borough_total.index.str[4:] + '-' + borough_total.index.str[0:4], format='%m-%Y')
    borough_total = borough_total.rename(columns={0:'Total thefts'})
    borough_chart = plot_theft_chart(borough_total)
    borough_chart.save('borough_charts/' + borough.replace(' ', '_') + '.html')
```
